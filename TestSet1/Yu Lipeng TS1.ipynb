```python
import pandas as pd
import numpy as np
```

```python
df_red = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv', sep=";")
df_white = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv', sep=";")
```

```python
#Red Wine Head
df_red.head()
```

```python
#White Wine Head
df_white.head()
```

```python
y1,y2 = (df_red['quality'].values > 6),(df_white['quality'].values > 6)
X1,X2 = df_red.iloc[:,0:-1],df_white.iloc[:,0:-1]
```

```python
from sklearn.cross_validation import train_test_split

X1_train, X1_test, y1_train, y1_test = \
    train_test_split(X1, y1, test_size=0.3, random_state=0)

X2_train, X2_test, y2_train, y2_test = \
    train_test_split(X2, y2, test_size=0.3, random_state=0)
```

```python
#Red Wine Description
from sklearn.preprocessing import StandardScaler

stdsc = StandardScaler()
X1_train_std = stdsc.fit_transform(X1_train)
X1_test_std = stdsc.transform(X1_test)
pd.DataFrame(X1_train_std).describe().transpose()
```

```python
#White Wine Descripton
stdsc = StandardScaler()
X2_train_std = stdsc.fit_transform(X2_train)
X2_test_std = stdsc.transform(X2_test)
pd.DataFrame(X2_train_std).describe().transpose()
```

```python
#Red Wine Linear Regression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,accuracy_score
lir = LinearRegression()
lir.fit(X1_train_std,y1_train)
lir1_train_pred,lir1_test_pred= lir.predict(X1_train_std),lir.predict(X1_test_std)

#Turn the regression result into 0 and 1
lir1_train_pred[lir1_train_pred >= 0.5],lir1_test_pred[lir1_test_pred >= 0.5] = 1,1
lir1_train_pred[lir1_train_pred < 0.5],lir1_test_pred[lir1_test_pred < 0.5] = 0,0

print('Red Wine Linear Regression\nTrain accuracy: %.3f; Test accuracy: %.3f' % (
accuracy_score(y1_train, lir1_train_pred),  
accuracy_score(y1_test, lir1_test_pred)))
```

```python
#White Wine Linear Regression
from sklearn.metrics import mean_squared_error,accuracy_score
lir = LinearRegression()
lir.fit(X2_train_std,y2_train)
lir2_train_pred,lir2_test_pred= lir.predict(X2_train_std),lir.predict(X2_test_std)

#Turn the regression result into 0 and 1
lir2_train_pred[lir2_train_pred >= 0.5],lir2_test_pred[lir2_test_pred >= 0.5] = 1,1
lir2_train_pred[lir2_train_pred < 0.5],lir2_test_pred[lir2_test_pred < 0.5] = 0,0

print('White Wine Linear Regression\nTrain accuracy: %.3f; Test accuracy: %.3f' % (
accuracy_score(y2_train, lir2_train_pred),  
accuracy_score(y2_test, lir2_test_pred)))
```

```python
#Red Wine Logistic Regression
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
import matplotlib.pyplot as plt

pipe_lr = Pipeline([('scl', StandardScaler()),
                    ('pca', PCA(n_components=2)),
                    ('clf', LogisticRegression(random_state=1))])

pipe_lr.fit(X1_train, y1_train)
print('Red Wine Logistic Regression\nTrain accuracy: %.3f; Test accuracy: %.3f' % (pipe_lr.score(X1_train, y1_train),pipe_lr.score(X1_test, y1_test)))
y1_pred = pipe_lr.predict(X1_test)
```

```python
#White Wine Logistic Regression
pipe_lr.fit(X2_train, y2_train)
print('White Wine Logistic Regression\nTrain accuracy: %.3f; Test accuracy: %.3f' % (pipe_lr.score(X2_train, y2_train),pipe_lr.score(X2_test, y2_test)))
y2_pred = pipe_lr.predict(X2_test)
```

```python
#Red Wine SVM
from sklearn.svm import SVC

pipe_lr = Pipeline([('scl', StandardScaler()),
                    ('pca', PCA(n_components=2)),
                    ('svm', SVC(kernel='rbf', random_state=0, gamma=0.20, C=5.0))])

pipe_lr.fit(X1_train, y1_train)
print('Red Wine SVM\nTrain accuracy: %.3f; Test accuracy: %.3f' % (pipe_lr.score(X1_train, y1_train),pipe_lr.score(X1_test, y1_test)))
y1_pred = pipe_lr.predict(X1_test)
```

```python
#White Wine SVM
pipe_lr.fit(X2_train, y2_train)
print('White Wine SVM\nTrain accuracy: %.3f; Test accuracy: %.3f' % (pipe_lr.score(X2_train, y2_train),pipe_lr.score(X2_test, y2_test)))
y2_pred = pipe_lr.predict(X2_test)
```
