```python
# Data processing
from sklearn.cross_validation import train_test_split
X, y = red_wine.iloc[:, :-1].values, red_wine['quality'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
X_train_norm1 = mms.fit_transform(X_train)
X_test_norm1 = mms.transform(X_test)

from sklearn.preprocessing import StandardScaler
stdsc = StandardScaler()
X_train_norm2 = stdsc.fit_transform(X_train)
X_test_norm2 = stdsc.transform(X_test)
```

```python
## Linear Regression
ols = LinearRegression()
ols.fit(X_train_norm2,y_train)
ols_train_pred = ols.predict(X_train_norm2)
ols_test_pred = ols.predict(X_test_norm2)
```

```python
dir(ols)
```

```python
ols.coef_
```

```python
print('OLS MSE train: %.3f, test: %.3f' % (
mean_squared_error(y_train, ols_train_pred),
mean_squared_error(y_test, ols_test_pred)))
```

```python
# Logistic regression
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train_norm2,y_train)
lr_train_pred = lr.predict(X_train_norm2)
lr_test_pred = lr.predict(X_test_norm2)

print('Logistic Regression MSE train: %.3f, test: %.3f' % (
mean_squared_error(y_train, lr_train_pred),
mean_squared_error(y_test, lr_test_pred)))
```

```python
# SVM
from sklearn.svm import SVC
svm = SVC(kernel='linear', C=1.0, random_state=0)

svm.fit(X_train_norm2,y_train)
svm_train_pred = svm.predict(X_train_norm2)
svm_test_pred = svm.predict(X_test_norm2)

print('SVM MSE train: %.3f, test: %.3f' % (
mean_squared_error(y_train, svm_train_pred),
mean_squared_error(y_test, svm_test_pred)))
```

```python
# SVM2
svm = SVC(kernel='rbf', C=1.0, random_state=0)

svm.fit(X_train_norm2,y_train)
svm_train_pred = svm.predict(X_train_norm2)
svm_test_pred = svm.predict(X_test_norm2)

print('SVM MSE2 train: %.3f, test: %.3f' % (
mean_squared_error(y_train, svm_train_pred),
mean_squared_error(y_test, svm_test_pred)))
```
