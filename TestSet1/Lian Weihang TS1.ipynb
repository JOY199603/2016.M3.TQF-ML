```python
import numpy as np
import pandas as pd
df_red = pd.read_csv('winequality-red.csv', sep=";")
df_white = pd.read_csv('winequality-white.csv', sep=";")
```

```python
y_red = (df_red['quality'].values > 6)
X_red = df_red.iloc[:,0:-1]
y_white=(df_white['quality'].values > 6)
X_white = df_white.iloc[:,0:-1]
```

```python
from sklearn.model_selection import train_test_split
X_red_train, X_red_test, y_red_train, y_red_test = \
    train_test_split(X_red, y_red, test_size=0.3, random_state=0)
X_white_train, X_white_test, y_white_train, y_white_test = \
    train_test_split(X_white, y_white, test_size=0.3, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
stdsc = StandardScaler()
X_red_train_std = stdsc.fit_transform(X_red_train)
X_red_test_std = stdsc.transform(X_red_test)
pd.DataFrame(X_red_train_std).describe().transpose()
```

```python
X_white_train_std = stdsc.fit_transform(X_white_train)
X_white_test_std = stdsc.transform(X_white_test)
pd.DataFrame(X_white_train_std).describe().transpose()
```

```python
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
pipe_lr = Pipeline([('scl', StandardScaler()),
                    ('pca', PCA(n_components=2)),
                    ('clf', LogisticRegression(random_state=1))])

pipe_lr.fit(X_red_train, y_red_train)
print('Red Wine logistic Rgression Test Accuracy: %.3f' % pipe_lr.score(X_red_test, y_red_test))
y_red_pred = pipe_lr.predict(X_red_test)
pipe_lr.fit(X_white_train, y_white_train)
print('White Wine logistic Rgression Test Accuracy: %.3f' % pipe_lr.score(X_white_test, y_white_test))
y_white_pred=pipe_lr.predict(X_white_test)
```

```python
from sklearn.svm import SVC
pipe_lr = Pipeline([('scl', StandardScaler()),
                    ('pca', PCA(n_components=2)),
                    ('svm', SVC(kernel='rbf', random_state=0, gamma=0.20, C=5.0))])
pipe_lr.fit(X_red_train, y_red_train)
print('Red Wine SVC Test Accuracy: %.3f' % pipe_lr.score(X_red_test, y_red_test))
y_red_pred = pipe_lr.predict(X_red_test)
pipe_lr.fit(X_white_train, y_white_train)
print('White Wine SVC Test Accuracy: %.3f' % pipe_lr.score(X_white_test, y_white_test))
y_white_pred=pipe_lr.predict(X_white_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,accuracy_score
lir = LinearRegression()
lir.fit(X_red_train_std,y_red_train)
lir_red_train_pred,lir_red_test_pred= lir.predict(X_red_train_std),lir.predict(X_red_test_std)
lir.fit(X_white_train_std,y_white_train)
lir_white_train_pred,lir_white_test_pred= lir.predict(X_white_train_std),lir.predict(X_white_test_std)
lir_red_train_pred[lir_red_train_pred >= 0.5],lir_red_test_pred[lir_red_test_pred >= 0.5] = 1,1
lir_red_train_pred[lir_red_train_pred < 0.5],lir_red_test_pred[lir_red_test_pred < 0.5] = 0,0
lir_white_train_pred[lir_white_train_pred >= 0.5],lir_white_test_pred[lir_white_test_pred >= 0.5] = 1,1
lir_white_train_pred[lir_white_train_pred < 0.5],lir_white_test_pred[lir_white_test_pred < 0.5] = 0,0
print('Red Wine Linear Regression Test accuracy: %.3f' % (accuracy_score(y_red_test, lir_red_test_pred)))
print('White Wine Linear Regression Test accuracy: %.3f' % (accuracy_score(y_white_test, lir_white_test_pred)))
```
