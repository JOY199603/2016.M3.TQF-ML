For red wine, Assing input (X) and output (y);
              Split training set and test set;
              Normalize X data (on training set).

```python
y = (df_red['quality'].values > 6)
X = df_red.iloc[:,0:-1]
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = \
    train_test_split(X, y, test_size=0.3, random_state=0)

from sklearn.preprocessing import StandardScaler
stdsc = StandardScaler()
X_train_std = stdsc.fit_transform(X_train)
X_test_std = stdsc.transform(X_test)
pd.DataFrame(X_train_std).describe().transpose()
```

For white wine, Assing input (N) and output (m);
              Split training set and test set;
              Normalize N data (on training set).

```python
m = (df_white['quality'].values > 6)
N = df_white.iloc[:,0:-1]
from sklearn.cross_validation import train_test_split
N_train, N_test, m_train, m_test = \
    train_test_split(N, m, test_size=0.3, random_state=0)

from sklearn.preprocessing import StandardScaler
stdsc = StandardScaler()
N_train_std = stdsc.fit_transform(N_train)
N_test_std = stdsc.transform(N_test)
pd.DataFrame(N_train_std).describe().transpose()
```

Apply normilization, PCA (first 2 components) and logistic regression on red wine.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
pipe_lr = Pipeline([('scl', StandardScaler()),
                    ('pca', PCA(n_components=2)),
                    ('clf', LogisticRegression(random_state=1))])
pipe_lr.fit(X_train, y_train)
print('Test Accuracy: %.3f' % pipe_lr.score(X_test, y_test))
y_pred = pipe_lr.predict(X_test)
```

Apply normilization, PCA (first 2 components) and logistic regression on white wine.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
pipe_lr = Pipeline([('scl', StandardScaler()),
                    ('pca', PCA(n_components=2)),
                    ('clf', LogisticRegression(random_state=1))])
pipe_lr.fit(N_train, m_train)
print('Test Accuracy: %.3f' % pipe_lr.score(N_test, m_test))
m_pred = pipe_lr.predict(N_test)
```

Apply Support Vector Machine with RBF kernel on red wine.

```python
from sklearn.svm import SVC

pipe_lr = Pipeline([('scl', StandardScaler()),
                    ('pca', PCA(n_components=2)),
                    ('svm', SVC(kernel='rbf', random_state=0, gamma=0.20, C=5.0))])

pipe_lr.fit(X_train, y_train)
print('Test Accuracy: %.3f' % pipe_lr.score(X_test, y_test))
y_pred = pipe_lr.predict(X_test)
```

Apply Support Vector Machine with RBF kernel on white wine.

```python
from sklearn.svm import SVC

pipe_lr = Pipeline([('scl', StandardScaler()),
                    ('pca', PCA(n_components=2)),
                    ('svm', SVC(kernel='rbf', random_state=0, gamma=0.20, C=5.0))])

pipe_lr.fit(N_train, m_train)
print('Test Accuracy: %.3f' % pipe_lr.score(N_test, m_test))
m_pred = pipe_lr.predict(N_test)
```

Linear Regression model of red wine and confusion matrix.

```python
import sklearn.linear_model as skl_lm

y = df_red['quality']

X_train, X_test, y_train, y_test = \
    train_test_split(X, y, test_size=0.3, random_state=0)

regr=skl_lm.LinearRegression()

regr.fit(X_train,y_train)

y_pred =regr.predict(X_test)

y_pred = (y_pred > 5.63)
y_test = (y_test > 5.63)
```

```python
from sklearn.metrics import confusion_matrix
outcome = confusion_matrix(y_true=y_test, y_pred=y_pred)
print(outcome)

result=np.matrix(outcome)
ERR=(result[0,1]+result[1,0])/(result[0,0]+result[0,1]+result[1,0]+result[1,1])
ERR
```

Linear Regression model of white wine and confusion matrix.

```python
import sklearn.linear_model as skl_lm

m = df_white['quality']

N_train, N_test, m_train, m_test = \
    train_test_split(N, m, test_size=0.3, random_state=0)

regr=skl_lm.LinearRegression()

regr.fit(N_train,m_train)

m_pred =regr.predict(N_test)

m_pred = (m_pred > 6)
m_test = (m_test > 6)
```
